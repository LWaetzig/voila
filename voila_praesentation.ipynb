## GENERAL INFORMATION

### What does Voilà do?
Voilà turns your jupyter notebook into a standalone website!

## How to install
using conda: mamba install -c conda-forge voila

using pip: pip install voila

### Usage
you can either use the add-on in jupyter notebook:


or run: voila {path-to-notebook} in you command shell to execute one specific notebook


if you want to observe a directory, just run voila:



### How does voila work?
When voila runs a notebook:
1. the code in the notebook is executed
2. the notebook and its outputs are converted to HTML. (by default, the notebook code cells are hidden)
3. The page is served e.g. via the Jupyter server
4. When users access the page, the widgets on the page have access to the underlying Jupyter kernal

### Customizing Voilà
There are many ways to customize your web-application and change the look and felling when interacting with it

#### Changing the theme
By default voila appears in a light theme. But luckily there is a way to change the default theme to dark by passing the following option when exectuing in the shell:

voila {path-to-notebook} --theme=dark



-> the theme can also be changed in the notebook metadata

#### use a template
- Voilà uses nbconvert to convert your notebook into an html dashboard
- nbconvert has a very rich templating system that allows you to customize the way in which your notebook is converted
- To use this template with voila either use the shell:

-> voila {path-to-template} --tempalte={name-of-the-template}

- or edit the metadata
