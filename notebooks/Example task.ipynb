```python
def interactiveplot(w=1):
    """Plots graph with that can be manipulated through Ipywidgets.

    Args:
        w (int, optional): Adds weight to Y-axis points. Defaults to 1.
    """
    x = X
    y = x*2+w*Y
    fig,axis = plt.subplots(1, figsize =(8,4))
    axis.scatter(x,y)
    
def histogramplot(mü=0, sigma=1, n=100, bins=10, color="gray"):
    """Plots dynamic histogram

    Args:
        mü (int, optional):  expected value. Defaults to 0.
        sigma (int, optional): variance. Defaults to 1.
        n (int, optional): number of blocks. Defaults to 100.
        bins (int, optional): width of blocks. Defaults to 10.
        color (str, optional): Defaults to "blue".
    """
    fig = plt.figure(figsize =(8,4))
    plt.xlim(-20,20)
    x = np.random.normal(mü, sigma, n)
    plt.hist(x, bins=bins, color=color)
    
def covidplot():
    """
    Plots figure that shows PS_COVID_Faelle to 2022-06-22 - 2020-03-02
    
    """
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=covid["Datum"], y=covid["PS_COVID_Faelle"]))
    fig.update_layout(title="Covid - Fälle Deutschland", xaxis_title="Datum", yaxis_title="Covid - Fälle")
    fig.show()

def rwertplot():
    """
    Plots figure that shows PS_7_Tage_R_Wert to 2022-06-22 - 2020-03-02
    
    """
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=covid["Datum"], y=covid["PS_7_Tage_R_Wert"]))
    fig.update_layout(title="R - Wert Deutschland", xaxis_title="Datum", yaxis_title="R - Wert")
    fig.show()
     
```

```python
widgets.interact(interactiveplot, w =(0, 50, 0.1));
```

```python
widgets.interact(histogramplot, mü=(-10,10, 0.5), sigma=(0,10,0.1), n=(10,1000,1),bins=(1,100,1), color=["gray", "green", "blue", "red"]);
```
